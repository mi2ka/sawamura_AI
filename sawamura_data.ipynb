## スクレイピングによりデータを取得する

プロ野球データFreak様（https://baseball-data.com/21/stats/pitcher-all/ip3-1.html）　を参考にした。

2009年から2021年の投手データが存在するので、それぞれをcsvファイルに変換して、一つのcsvファイルに結合する。

まとめたファイルを「Pitcher_total.csv」 とする。

このままだと”沢村賞”、年度を区別するための”year”、”完投数”、”完封数”という項目がないため、Excelを使用して適宜追加する。

Excelで処理を行ったファイルを「SawamuraPitcher_total.csv」とする。

完投数と完封数はnf3-Baseball Data House-様を参考にした。（https://nf3.sakura.ne.jp/php/stat_disp/stat_disp.php?y=0&leg=1&tm=Sta&fp=1&dn=6&dk=0%27　）

”沢村賞”は受賞者を「１」としそれ以外の全投手を「０」とする。

↓以下は2021年から2009年の投手データを取得する処理と各々のファイルを一つにまとめる処理

```python
import urllib
#HTMLからデータを抽出するために使用
from bs4 import BeautifulSoup
#表形式でデータ格納するために使用
import pandas as pd

#urlが変更された
#url = 'https://baseball-data.com/stats/pitcher-all/ip3-1.html'

url = 'https://baseball-data.com/21/stats/pitcher-all/ip3-1.html'
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
 
#HTMLから表(tableタグ)の部分を全て取得する
table = soup.find_all('table')[0]
#テーブルから行データを取得
rows = table.find_all('tr')
 
#データ格納用
headData = []
playerData = []
    
for i, row in enumerate(rows):    
    #1行目
    if i == 0:
        #ヘッダ部取得
        for headerValue in row.find_all('th'):
            headData.append(headerValue.get_text())
    #2行目～最終行未満
    elif (i > 0) and (i + 1 < len(rows)):
        #明細部取得
        playerRow = []
        for playerValue in row.find_all('td'):
            playerRow.append(playerValue.get_text())
        playerData.append(playerRow)
 
#DataFrameに変換
df = pd.DataFrame(data = playerData, columns = headData)
df["year"] = str(20) + str(21)
print(df)
 
#CSV出力
df.to_csv('PitcherData_21.csv', sep = ',', header = True, index = False)
```

## csvファイル結合

```python
import glob

# パスで指定したファイルの一覧をリスト形式で取得. 
csv_files = glob.glob('sawamura/*.csv')

#読み込むファイルのリストを表示
for a in csv_files:
    print(a)

#csvファイルの中身を追加していくリストを用意
data_list = []

#読み込むファイルのリストを走査
for file in csv_files:
    data_list.append(pd.read_csv(file))

#リストを全て行方向に結合
#axis=0:行方向に結合, sort
df = pd.concat(data_list, axis=0)
df.to_csv('Pitcher_total.csv',index=False)
```

作成したPitcher_total.csvをExcelを使用して件の処理を行った。そしてそのファイルをSawamuraPitcher_total.csvと名付けた。
